### Importing useful library

```python
import pandas as pd
from datetime import date
from nsepy import get_history
```

### Getting TCS data

```python
tcs = get_history(symbol='TCS',
                   start=date(2015,1,1),
                   end=date(2015,12,31))
```

```python
tcs.head(10)
```

### Getiing INFY data

```python
infy = get_history(symbol='INFY',
                   start=date(2015,1,1),
                   end=date(2015,12,31))
```

```python
infy.head(5)
```

### Getting NIFTY IT data

```python
nifty_it = get_history(symbol="NIFTYIT",
                            start=date(2015,1,1),
                            end=date(2015,12,31),
                            index=True)
```

```python
nifty_it.head(5)
```

### Saving to CSV file 

```python
tcs.to_csv('tcs_stock.csv', encoding='utf-8')
infy.to_csv('infy_stock.csv', encoding='utf-8')
nifty_it.to_csv('nifty_it_index.csv', encoding='utf-8')
```
